##### Binary search is a technique to reduce the search space by half at every step using a decision rule. It can be used whenever the search space is monotonic(A condition is monotonic if it changes only once in one direction) like ex: True  True  True  False False False --> Invalid example True False True
###### Binary Search is NOT just for sorted arrays. Sorted arrays are just one special case ----> Mental Checklist to Identify Binary Search
###### 1. Can I discard half the search space? 2. Does the condition behave in one direction only? 3. Is there a single boundary where behavior changes?

##### There are three methods in Binary search(It's also known as bisection search is an algorithm for finding an element quickly in a sorted Array)
###### 1. Traditional method(Looking for the element if it exists in the sorted list/array)
###### 2. Overlook(skipping over the answer). Main cause is using mid+1 when not needed
###### 3. Underlook(failing to reach the half). Main cause is using mid when you should use mid+1

#### Concrete Example of Integer Overflow in (C, C++, Java) in Binary Search. This doesn't happend in pyhton.
##### Assume we are working in a 32-bit signed integer environment (C, C++, Java). Max int = 2,147,483,647  (‚âà 2.1 billion). Now run binary search on a very large array:
###### int LOW  = 2,000,000,000;   // 2 billion int HIGH = 2,100,000,000;   // 2.1 billion
###### int MID = (LOW + HIGH) / 2; LOW + HIGH = 2,000,000,000 + 2,100,000,000 = 4,100,000,000 --> OVERFLOW (cannot fit in 32-bit int)
###### Correct Version --> int MID = LOW + (HIGH - LOW) / 2;
##### But in python 4,100,000,000 number fits perfectly fine because Python automatically expands integer size to fit any value. Python does not use 32-bit or 64-bit integers internally. It dynamically allocates more memory as needed.

```python
# https://leetcode.com/problems/peak-index-in-a-mountain-array/
class Solution:
    def peakIndexInMountainArray(self, arr: List[int]) -> int:
        # i = 0
        # while arr[i]<arr[i+1]:
        #     i +=1
        # return i
        left, right = 0, len(arr)-1
        while left < right:
            mid = (left+right)//2
            if arr[mid]< arr[mid+1]:
                left = mid+1
            else:
                right = mid
        return left
```

```python
# https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/
class Solution:
    def findMin(self, nums: List[int]) -> int:
        l, r = 0, len(nums)-1
        while l<r:
            m = (l+r)//2
            if nums[m]>nums[r]:
                l = m+1
            else:
                r = m
        return nums[l]
```

##### l < r vs l <= r
##### Use while l <= r üëâ When you are searching for an element --> You want to check every possible index, including when l == r.
##### Use while l < r üëâ When you are searching for a boundary / answer --> You are not checking for equality. You are letting pointers converge to the answer.

```python
# https://leetcode.com/problems/search-in-rotated-sorted-array-ii/
# SAME AS ABOVE BUT USED 3 STEP PROCESS
class Solution:
    def search(self, nums: List[int], target: int) -> bool:
        n = len(nums)
        l, r = 0, n - 1
        # STEP 1: Find the index of the minimum element (pivot point)
        while l < r:
            m = (l + r) // 2
            if nums[m] > nums[r]:
                l = m + 1
            elif nums[m] < nums[r]:
                r = m
            else:
                # shrink duplicates on BOTH ends
                while l < r and nums[r - 1] == nums[r]:
                    r -= 1
                while l < r and nums[l + 1] == nums[l]:
                    l += 1
        min_index = l
        # STEP 2: Determine which sorted half the target belongs to
        if min_index == 0:  # Case 1: No rotation, entire array sorted normally
            l, r = 0, n - 1
        # Case 2: Target lies within the left sorted portion
        elif nums[0] <= target <= nums[min_index - 1]:
            l, r = 0, min_index - 1
        # Case 3: Target lies within the right sorted portion
        else:
            l, r = min_index, n - 1
        # STEP 3: Standard binary search on the selected half
        while l <= r:
            mid = (l + r) // 2
            if nums[mid] == target:
                return True
            elif nums[mid] < target:
                l = mid + 1
            else:
                r = mid - 1
        return False
```

```python
# https://leetcode.com/problems/search-a-2d-matrix-ii/
# We can Solve this in binary search by searching diagonal + binary search. But it is hard and simple method is top-right and down elimination
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        m = len(matrix)
        n = len(matrix[0])
        # STEP 1: Start at top-right and move inside the matrix
        row, col = 0, n-1
        # STEP 2: Move left or down based on comparison
        while row<m and col>=0:
            if matrix[row][col] == target:
                return True
            elif matrix[row][col]>target:
                col -=1 # move left
            else:
                row +=1 # move down
        return False
```

```python
class Solution:
    def findClosestElements(self, arr: List[int], k: int, x: int) -> List[int]:
        # BRUTFORCE O(N LOG N)
        # arr = [(abs(arr[i]-x), arr[i]) for i in range(len(arr))]
        # arr.sort()
        # res = []
        # for i in range(k):
        #     res.append(arr[i][1])
        # return sorted(res)
        l, r = 0, len(arr)-k # We set r = n - k to avoid starting a window that doesn‚Äôt have enough elements.
        while l<r:
            m = (l+r)//2
            # If x - arr[mid] is greater than arr[mid + k] - x, that means: x is closer to the right side than to the left side of this window or vice-versa
            if x-arr[m]>arr[m+k]-x:
                l = m+1
            else:
                r = m
        return arr[l:l+k]
```

```python
# https://leetcode.com/problems/koko-eating-bananas/
import math
def minEatingSpeed(self, piles: List[int], h: int) -> int:
        # BRUTEFORCE O(N*M) N = Array size, M = Max of bananas in a single pile from piles
        # speed = 1
        # while True:
        #     hour_spent = 0
        #     for pile in piles:
        #         hour_spent +=math.ceil(pile/speed)
        #     if hour_spent <= h:
        #         return speed
        #     else:
        #         speed += 1
        left = 1
        right = max(piles)
        while left < right:
            hour_spent = 0
            mid = (left + right)//2
            for pile in piles:
                hour_spent += math.ceil(pile/mid)
            if hour_spent <= h:
                right = mid
            else:
                left = mid+1
        return left
```
