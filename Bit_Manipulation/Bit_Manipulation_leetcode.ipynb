```python
# https://leetcode.com/problems/single-number/
def singleNumber(self, nums: List[int]) -> int:
        res  = 0
        for num in nums:
            res ^=num
        return res
```

```python
# https://leetcode.com/problems/number-of-1-bits/
def hammingWeight(n: int) -> int:
    # # This is Not ideal way because we don't need to convert number to binary again. becuase on computer number is stored as binary alreday
    # number_to_binary = bin(n)[2:]
    # count = 0
    # for i in number_to_binary:
    #     if i == "1":
    #         count+=1
    # return count
    # # Short Division by Two with Remainder. Time is O(log n) but below method is fastest.
    # ans=0
    # while n>0:
    #     if n%2==1:
    #         ans+=1
    #     n=n/2
    # return ans
    count = 0
    while n:
        n &= (n-1)
        count +=1
    return count
```

```python
# https://leetcode.com/problems/counting-bits/
def countBits(self, n: int) -> List[int]:
    # Time is O(NLogN) but can be done in O(N) Using DP.
    ans = []
    for num in range(n+1):
        count = 0
        while num:
            num &= (num-1)
            count +=1
        ans.append(count)
    return ans
```

```python
# https://leetcode.com/problems/missing-number/
def missingNumber(self, nums: List[int]) -> int:
    # BRUTEFORCE SOLUTION O(N^2) time
    # for i in range(n+1):
    #     if i not in nums:
    #         return i
    # return None

    # # MATH SOLUTION --> O(N)
    # n = len(nums)
    # sum_n_numbers = (n*(n+1))//2
    # missing_number = sum_n_numbers - sum(nums)
    # return missing_number
    
    # BITWISE XOR solution --> O(N)
    n = len(nums)
    ans = n
    for i in range(n):
        ans ^= i ^ nums[i]
    return ans
```

```python
# https://leetcode.com/problems/sum-of-two-integers/
def getSum(self, a: int, b: int) -> int:
    mask = 0xffffffff
    max_int = 0x7fffffff
    while b!=0:
        sum_ = (a^b) & mask
        carry = ((a&b) << 1) & mask
        a,b = sum_, carry
    return a if a<=max_int else ~(a^mask)
```

```python
# https://leetcode.com/problems/reverse-integer/
def reverse(self, x: int) -> int:
    sign = -1 if x < 0 else 1
    x = abs(x)
    rev = 0
    while x != 0:
        digit = x % 10
        rev = rev * 10 + digit
        x //= 10
    if rev < -2**31 or rev > 2**31 - 1:
        return 0
    return rev*sign
```
