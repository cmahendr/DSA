###### Add/Pop takes O(log n), peek(top element) O(1), heapify takes O(n), Heap Sort takes O(n*log n) because we pop min element and it takes O(lon n) + we do it for n elements so O(n * log n)
##### Import heapqfor using Heaps in Python. heapq only support min heap and to do max heap 
###### Push negative values (Most common / easiest) This is the standard technique used everywhere (LeetCode, interviews, NeetCode).
###### ✔ Min-heap + negative values = Max-heap Because:
###### Largest number → smallest negative number
###### Smallest number → largest negative number

```python
import heapq
# IF THERE ARE NO ELEMENTS IN LIST AND TRYING TO PUSH ELE AND WANT TO WRITE THE MAX ELE
max_heap = []
# push value v as -v
heapq.heappush(max_heap, -5)
heapq.heappush(max_heap, -6)
heapq.heappush(max_heap, -3)
print(max_heap)
# to get max value: take negative again
print(-heapq.heappop(max_heap))  # 6
print(-heapq.heappop(max_heap))  # 5
print(-heapq.heappop(max_heap))  # 3
```

```python
# 1. What if the list already has positive numbers?
# 2. Convert whole list to max-heap by negating everything
arr = [10]  # existing list
max_heap = [-x for x in arr]  # convert to negative
heapq.heapify(max_heap)

# now push more (negated)
heapq.heappush(max_heap, -5)
heapq.heappush(max_heap, -6)
heapq.heappush(max_heap, -3)
print(max_heap)

# get max element:
print(-heapq.heappop(max_heap))  # 10
```

```python
import heapq
class KthLargest:
    def __init__(self, k: int, nums: List[int]):
        self.k, self.minHeap = k, nums
        heapq.heapify(self.minHeap)
        while len(self.minHeap) > k:
            heapq.heappop(self.minHeap)

    def add(self, val: int) -> int:
        heapq.heappush(self.minHeap, val)
        if len(self.minHeap) > self.k:
            heapq.heappop(self.minHeap)
        return self.minHeap[0]
```

```python
# https://leetcode.com/problems/kth-smallest-element-in-a-sorted-matrix/
# MIN-HEAP SOLUTION MUCH BETTER THAN BRUTFORCE BUT GO WITH BINARY SEARCH METHOD IF THE K VALUE IS BIG we can also use max_heap but not good compared to min heap and binary search
import heapq
class Solution:
    def kthSmallest(self, matrix: List[List[int]], k: int) -> int:
        minHeap = []
        n = len(matrix)
        m = min(k,n)
        for row in range(m):
            heapq.heappush(minHeap, (matrix[row][0], row, 0))
        while k:
            val, row, col = heapq.heappop(minHeap)
            if col < n-1:
                heapq.heappush(minHeap, (matrix[row][col+1], row, col+1))
            k -=1
        return val

# BINARY SEARCH SOLUTION
class Solution:
    def kthSmallest(self, matrix: List[List[int]], k: int) -> int:
        n = len(matrix)
        start, end = matrix[0][0], matrix[n - 1][n - 1]
        while start < end:
            mid = (start + end) // 2
            count = self.count(matrix, mid)
            if count < k:
                start = mid+1  # kth smallest is larger
            else:
                end = mid  # kth smallest is <= mid
        return start
    def count(self, matrix, mid):
        count, n = 0, len(matrix)
        row, col = 0, len(matrix[0])-1
        while row < n and col >= 0:
            if matrix[row][col] > mid:
                col -= 1
            else:
                count += col + 1
                row += 1
        return count
```

```python
# https://leetcode.com/problems/reorganize-string/
from collections import Counter
class Solution:
    def reorganizeString(self, s: str) -> str:
        counter = Counter(s)
        maxHeap = [(-count, char) for char, count in counter.items()]
        heapq.heapify(maxHeap)
        cooldown = deque()
        time = 0
        res = []
        while maxHeap or cooldown:
            time +=1
            if maxHeap:
                count, char = heapq.heappop(maxHeap)
                res.append(char)
                count +=1
                if count != 0:
                    cooldown.append((count, time+1, char))
            else:
                return ''
            if cooldown and cooldown[0][1] == time:
                count, _, char = cooldown.popleft()
                heapq.heappush(maxHeap, (count, char))
        return ''.join(res)
```

```python
# https://leetcode.com/problems/find-k-pairs-with-smallest-sums/
import heapq
class Solution:
    def kSmallestPairs(self, nums1, nums2, k):
        if not nums1 or not nums2:
            return []
        minHeap = []
        res = []
        for i in range(min(k, len(nums1))):
            heapq.heappush(minHeap, (nums1[i] + nums2[0], i, 0))
        while k and minHeap:
            _, i, j = heapq.heappop(minHeap)
            res.append([nums1[i], nums2[j]])
            if j+1 < len(nums2):
                heapq.heappush(minHeap, (nums1[i] + nums2[j+1], i, j+1))
            k -=1
        return res
```
