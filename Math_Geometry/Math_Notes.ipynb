```python
def amstrong(x):
    res  = 0
    while x>0:
        res += (x%10)**3 # in Python ** is the exponentiation operator BUT, "^"" is the bitwise XOR operator. So, use ** for exponentiation.
        x = x//10
    return res
print(amstrong(153)==143)
```

```python
def divisors(x):
    small  = []
    larger = []
    for i in range(1, int(x**0.5)+1): # we only need to check up to the square root of x, because if x is divisible by any number greater than its square root, 
                                        # it must have a corresponding divisor that is smaller than the square root.
        if x % i ==0:
            small.append(i)
            if i != x//i: # to avoid adding the square root twice in case of perfect squares --> We can also write x//i not in small, but current is more efficient because it avoids the need to search through the list.
                larger.append(x//i)
    return small + larger[::-1]

print(divisors(36))
```

```python
def prime(x):
    # if x<2:
    #     return False
    # for i in range(2, int(x**0.5)+1):
    #     if x % i == 0:
    #         return False
    # return True
    res  = []
    for i in range(2, x+1): # Time complexity of this function is O(n*sqrt(n)) because we are checking each number from 2 to n and for each number we are checking if it is prime by iterating up to its square root.
        # To optimize this function, we can use the Sieve of Eratosthenes algorithm which has a time complexity of O(n log log n) and uses O(n) space.
        is_prime = True
        for j in range(2, int(i**0.5)+1): # This gives Time complexity of O(sqrt(n)) for each number i, and since we are doing this for all numbers from 2 to n, the overall time complexity is O(n*sqrt(n)).
            # If i use range (2, i) instead of range(2, int(i**0.5)+1), then the time complexity would be O(n^2) because we would be checking all numbers less than i for each i, which is inefficient.
            if i % j == 0:
                is_prime = False
                break
        if is_prime:
            res.append(i)
    return res
print(prime(30))
```

```python
def sieve(n): # Time complexity of this function is O(n log log n) and space complexity is O(n) because we are creating a boolean array of size n+1 to keep track of prime numbers.
    # Step 1: Create a boolean array
    is_prime = [1] * (n + 1)
    is_prime[0] = is_prime[1] = 0
    # Step 2: Sieve process
    i = 2
    while i * i <= n:
        if is_prime[i]:
            # Mark multiples of i as not prime
            for j in range(i * i, n + 1, i):
                is_prime[j] = 0
        i += 1
    # Step 3: Collect primes
    primes = []
    for i in range(2, n + 1):
        if is_prime[i]:
            primes.append(i)
    return primes
print(sieve(15))
```

```python
def prime_factors(n):
    res  = []
    # for i in range(2, n+1): # Time is O(n) because in the worst case, when x is a prime number, we will check all numbers from 2 to x.
    #     while n% i ==0:
    #         res.append(i)
    #         n = n//i
    # return res
    # Check for number of 2s that divide n
    while n%2==0:
        res.append(2)
        n //= 2
    # n must be odd at this point, so we can skip even numbers
    for i in range(3, int(n**0.5)+1, 2):
        while n%i == 0:
            res.append(i)
            n //= i
    # This condition is to check if n is a prime number greater than 2, because if n is a prime number greater than 2, then it will not be reduced to 1 by the above loops and we need to add it to the result list.
    if n>2:
        res.append(n)
    return res # Time complexity of this function is O(sqrt(n)) in the worst case, when n is a prime number, because we will check all odd numbers up to the square root of n. In the best case, when n is a power of 2, the time complexity is O(log n) because we will keep dividing n by 2 until it becomes 1.
print(prime_factors(200))
```

```python
# Smallest Prime Factor (SPF) | Prime Factorisation | Query Based Problem
def prime_factors_bruteforce(x):
    res = []
    # divide by 2
    while x % 2 == 0:
        res.append(2)
        x //= 2
    # divide by odd numbers
    i = 3
    while i * i <= x:
        while x % i == 0:
            res.append(i)
            x //= i
        i += 2
    # leftover prime
    if x > 1:
        res.append(x)
    return res
# Example queries
queries = [60, 84, 97, 100] # Time O(q.sqrt(n)) where q is the number of queries and n is the maximum number in the queries.
for x in queries:
    print(x, "->", prime_factors_bruteforce(x))
```

```python
def build_spf(n):
    spf = [0] * (n + 1)
    for i in range(2, n + 1):
        spf[i] = i
    for i in range(2, int(n**0.5) + 1):
        if spf[i] == i:  # prime
            for j in range(i * i, n + 1, i):
                if spf[j] == j:
                    spf[j] = i
    return spf
def prime_factors_using_spf(x, spf):
    res = []
    while x != 1:
        res.append(spf[x])
        x //= spf[x]
    return res
# Example
queries = [60, 84, 97, 100] # Time O(N log log N + q log N) where q is the number of queries and n is the maximum number in the queries, 
num = max(queries)
spf = build_spf(num)

# because for each query we are dividing x by its smallest prime factor until it becomes 1, and in the worst case, when x is a prime number, we will have log n divisions.
for x in queries:
    print(x, "->", prime_factors_using_spf(x, spf))

# Why SPF is better for many queries
# With SPF: Precompute once --> Factorization per query becomes O(log n)
```

```python
# Count Prime in a range L-R with q queries
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def count_primes_bruteforce(queries):
    result = []
    for L, R in queries:
        primes = []
        count = 0
        for num in range(L, R + 1):
            if is_prime(num):
                primes.append(num)
                count += 1
        result.append((count, primes))
    return result

# Example
queries = [(1, 10), (10, 30), (1, 100)]
print(count_primes_bruteforce(queries)) # Time complexity of this function is O(q * n * sqrt(n)) where q is the number of queries and n is the size of the range in each query. 
# This is because for each query, we are iterating through the range and checking if each number is prime, which takes O(sqrt(n)) time. Space complexity is O(p) where p is the number of prime numbers found in the ranges of the queries, because we are storing the prime numbers in a list.
```

```python
# Optimized version using Sieve of Eratosthenes
def sieve(n):
    is_prime = [1] * (n + 1)
    is_prime[0] = is_prime[1] = 0
    i = 2
    while i * i <= n:
        if is_prime[i]:
            for j in range(i * i, n + 1, i):
                is_prime[j] = 0
        i += 1
    return is_prime

def build_prefix(is_prime):
    prefix = [0] * len(is_prime)
    for i in range(1, len(is_prime)):
        prefix[i] = prefix[i - 1] + is_prime[i]
    return prefix

def count_primes_optimized(queries):
    max_r = max(R for _, R in queries)
    is_prime = sieve(max_r)
    prefix = build_prefix(is_prime)
    result = []
    for L, R in queries:
        result.append(prefix[R] - prefix[L - 1])
    return result
# Example
queries = [(1, 10), (10, 30), (1, 100)]
print(count_primes_optimized(queries))

```

```python
# LCM of an Array (Multiple Numbers)
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
def lcm(a, b):
    return (a * b) // gcd(a, b)
def lcm_array(arr):
    ans = arr[0]
    for i in range(1, len(arr)):
        ans = lcm(ans, arr[i])
    return ans

print(lcm_array([2, 6, 8, 14]))
```

```python
def power(x, n): # Time complexity is O(log n) because we are halving n at each step. 
                # Space complexity is O(1) because we are using only a constant amount of extra space to store the result and intermediate values.
    # ans = 1
    # if n<0:
    #     x = 1/x
    #     n =-n
    # while n>0:
    #     if n%2 ==1:
    #         ans *=x
    #         n -=1
    #     else:
    #         x= x*x
    #         n//=2
    # return ans
    if n<0:
        return 1/power(x, -n) # we can also write power(1/x, -n) but it is less efficient because it requires an additional recursive call to compute 1/x.
    if n == 0:
        return 1
    elif n == 1:
        return x
    elif n%2 == 0:
        return  power(x*x, n//2) # we can also write power(x, n//2) * power(x, n//2) which is less efficient because it requires two recursive calls instead of one.
    else:
        return x * power(x*x, n//2) # we can also write x * power(x, n//2) * power(x, n//2) which is less efficient because it requires two recursive calls instead of one.
    # Time complexity is O(log n) because we are halving n at each step. Recursive calls are made only for n//2, which reduces the number of calls significantly compared to a linear approach.
    # Total Time complexity is O(log n) and Space complexity is O(log n) due to the recursive call stack. However, if we use an iterative approach, we can achieve O(1) space complexity.
print(power(2,-2))
```

```python
def factorial(n):
    ans  = 1
    while n > 1:
        ans *= n
        n -= 1
    return ans
print(factorial(5))  # 120

def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)
print(factorial_recursive(5))  # 120
```

```python
def plusOne(digits: List[int]) -> List[int]:
    n =len(digits)
    for i in range(n-1, -1, -1):
        if digits[i] < 9:
            digits[i] +=1
            return digits
        digits[i] = 0
    return [1] + digits
```

#### Permutations: (nPr) = (n!)/(n-r)!
##### Arranging things in different order. For permutations order matters(Means that (A,B) â‰  (B,A).). Permutations: Used when sequence is important(e.g., passwords, rankings, race results).
###### Example with A, B, C (taking 2 letters at a time)
###### If order matters (permutations):AB and BA are different. So the set is: {AB, AC, BA, BC, CA, CB} â†’ 6 permutations.

#### Combinations: (nCr) = (n!)/(n-r)!*r! OR (nPr)/r!
##### Selection of things where order doesn't matter. Youâ€™re only choosing the group, not arranging it.
###### Example: If order doesnâ€™t matter (combinations): AB and BA are treated as the same group. So the set is: {AB, AC, BC} â†’ 3 combinations.
###### Combinations: Used when only the selection matters (e.g., choosing a team, lottery numbers, committee members).
