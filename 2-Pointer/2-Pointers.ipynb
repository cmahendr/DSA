```python
# https://leetcode.com/problems/valid-palindrome/
def isPalindrome(self, s: str) -> bool:
        # res = ''.join(ch.lower() for ch in s if ch.isalnum())
        # for i in range(len(res)):
        #     if res[i] != res[-i-1]:
        #         return False
        # return True
        left, right = 0, len(s)-1
        while left < right:
            if not s[left].isalnum():
                left +=1
                continue
            if not s[right].isalnum():
                right -=1
                continue
            if s[left].lower() != s[right].lower():
                return False
            left +=1
            right -=1
        return True
```

```python
# https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/
def twoSum(self, numbers: List[int], target: int) -> List[int]:
        # target_num_diff_dict = {}
        # for idx, num in enumerate(numbers):
        #     diff = target - num
        #     if num in target_num_diff_dict:
        #         return [target_num_diff_dict[num], idx+1]
        #     target_num_diff_dict[diff] = idx+1
        # return []
        # CONSTANT SPACE 
        left, right = 0, len(numbers)-1
        while left<right:
            curr_sum = numbers[left] + numbers[right]
            if curr_sum > target:
                right -=1
            elif curr_sum < target:
                left +=1
            else:
                return [left+1, right+1]
```

```python
# https://leetcode.com/problems/3sum/
def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        res  = []
        n  = len(nums)
        for i in range(n-2):
            # Because when i reaches n-2 or n-1, there aren’t enough numbers left to form a triplet, 
            # so those iterations do nothing and are wasted work, which is why we stop at n-2 instead of looping to n
            if i>0 and nums[i] == nums[i-1]:
                continue
            left, right = i+1, n-1
            while left < right:
                total = nums[left]+nums[right]+nums[i]
                if total == 0:
                    res.append([nums[i], nums[left], nums[right]])
                    left +=1
                    right -=1
                    while left < right and nums[left] == nums[left-1]:
                        left +=1
                    while left < right and nums[right] == nums[right+1]:
                        right -=1
                elif total<0:
                    left +=1
                else:
                    right -=1
        return res
```

```python
# https://leetcode.com/problems/3sum-closest/
def threeSumClosest(self, nums: List[int], target: int) -> int:
        # -4,-1,1,2
        nums.sort()
        n = len(nums)
        closest_sum = nums[0] + nums[1] + nums[2] #-4
        for i in range(n-2):
            if i>0 and nums[i] == nums[i-1]:
                continue
            left, right = i+1, n-1
            while left<right:
                curr_sum = nums[i]+nums[left]+nums[right] #-3
                if abs(curr_sum - target) < abs(closest_sum - target):
                    closest_sum = curr_sum
                if curr_sum==target:
                    return curr_sum
                elif curr_sum<target:
                    left +=1
                else:
                    right -=1
        return closest_sum
```

```python
# https://leetcode.com/problems/container-with-most-water/description/
def maxArea(self, height: List[int]) -> int:
        n = len(height)
        max_area = 0
        left, right = 0, n-1
        while left<right:
            width = right-left
            curr_area = min(height[left], height[right]) * width
            max_area = max(max_area, curr_area)
            if height[left]<=height[right]:
                left +=1
            else:
                right-=1
        return max_area
```

```python
# https://leetcode.com/problems/sort-colors/
def sortColors(self, nums: List[int]) -> None:
        left = curr = 0
        right = len(nums)-1
        while curr<=right:
            if nums[curr] == 0:
                nums[left], nums[curr] = nums[curr], nums[left]
                left +=1
                curr +=1
            elif nums[curr] == 2:
                nums[curr], nums[right] = nums[right], nums[curr]
                right -=1
            else:
                curr += 1
```
