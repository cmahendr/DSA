```python
# https://leetcode.com/problems/two-sum/
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        dict = {}
        for idx, val in enumerate(nums):
            diff = target - val
            if val in dict:
                return [dict[val], idx]
            dict[diff] = idx
        return []
```

```python
# https://leetcode.com/problems/group-anagrams/description/
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        # # O(NKLOG K) BECAUSE WE ARE SORTING
        # groups = {}
        # for word in strs:
        #     freq = tuple(sorted(Counter(word).items()))
        #     if freq not in groups:
        #         groups[freq] = []
        #     groups[freq].append(word)
        # return list(groups.values())
        groups = {}
        for word in strs:
            count = [0]*26
            for ch in word:
                count[ord(ch)- ord('a')] +=1
            # In python dict keys must be immutable so we can't use COUNT LIST Instead we used TUPLE.
            key = tuple(count)
            if key not in groups:
                groups[key] = []
            groups[key].append(word)
        return list(groups.values())
```

```python
# https://leetcode.com/problems/encode-and-decode-strings/
class Codec:
    def encode(self, strs: List[str]) -> str:
        """Encodes a list of strings to a single string.
        """
        # return 'π'.join(strs)
        res = ""
        for s in strs:
            res += str(len(s)) + "#" + s
        return res

    def decode(self, s: str) -> List[str]:
        """Decodes a single string to a list of strings.
        """
        # return s.split('π')
        res, i = [], 0
        while i < len(s):
            j = i
            while s[j]!= "#":
                j +=1
            length = int(s[i:j])
            res.append(str(s[j+1:j+1+length]))
            i = j+1+length
        return res
```

```python
# https://leetcode.com/problems/product-of-array-except-self/
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n =len(nums)
        res = [1] * n
        left = 1
        for i in range(n):
            res[i] = left
            left *=nums[i]
        right = 1
        for i in range(n-1, -1, -1):
            res[i] *= right
            right *= nums[i]
        return res
```

```python
# https://leetcode.com/problems/find-all-duplicates-in-an-array/
class Solution:
    def findDuplicates(self, nums: List[int]) -> List[int]:
        # USE SETS and COUNTER(WHICH USES HASHMAP) for solving the problem BUT AUXILARY MEMORY USED WILL BE O(N). WE NEED O(1) AUXILARY MEMORY
        res = []
        for num in nums:
            idx = abs(num) - 1
            if nums[idx]<0:
                res.append(abs(num))
            else:
                nums[idx] *= -1
        return res
```

#### Pre-Fix Sums Technique

#### Sliding window fails with negative numbers because multiple valid continous subarrays can end at the same index, but the technique maintains only one active window.
###### Input: [1,-1,1] and k=1 and ouput = 3 as follows [1] (index 0)| [1, -1, 1]  (index 0 → 2)| [1] (index 2) but 2-pointer(sliding window) gives only two --> [1] (index 0)| [1, -1, 1]  (index 0 → 2) misses [1] (index 2). So to solve this use pre-fix sums technique


```python
# https://leetcode.com/problems/subarray-sum-equals-k/description/
def subarraySum(self, nums, k):
        prefix = {0:1}
        count = 0
        curSum = 0
        for num in nums:
            curSum += num
            diff = curSum - k
            if diff in prefix:
                count += prefix[diff]
            prefix[curSum] = prefix.get(curSum, 0)+1
        return count
```

```python
# https://leetcode.com/problems/subarray-sums-divisible-by-k/
def subarraysDivByK(self, nums: List[int], k: int) -> int:
        prefix  = {0:1}
        currSum = 0
        count = 0
        for num in nums:
            currSum = currSum + num
            currSumMod =  (currSum)%k
            if currSumMod in prefix:
                count += prefix[currSumMod]
            prefix[currSumMod] = prefix.get(currSumMod, 0)+1
        return count
```
